## Serpinski Triangle from Randomness

Idea: have 3 points which create an equalateral triangle and find a random point within this triangle. Then, randomly pick one of the 3 points and move halfway to that point. This is the new origin point. Itterate this 'random walk' over time and analyze the overarching structure.

The goal of this program is to show how uniform structures can emerge from randomized variables, i.e. uniformality from chaos.

```python
import numpy as np
import matplotlib.pyplot as plt

# creating the triangle

P_X = [0,10,5]
P_Y = [0,0,9.238795325112868]

fig = plt.figure(figsize=[10,8])
ax = fig.add_subplot(111)
ax.plot(P_X,P_Y,'ko')

# make the first point (P_0)
P_0x = np.random.randint(0,11)
P_0y = np.random.randint(0,11)
print(P_0x,P_0y)
ax.plot(P_0x,P_0y,'bo')

# Making the iterable
iters = 100000
cnt = 1

f = open("Serpinski_data.txt", 'w')

for i in range(iters):
#  print("count = %1.0f" %cnt)
  pick = np.random.randint(0,3)
#  print("pick = %1.0f" %pick)
  if pick == 0:
    P_0x = (P_0x - P_X[0])/2
    P_0y = (P_0y - P_Y[0])/2
  if pick == 1:
    P_0x = (P_0x - P_X[1])/2
    P_0y = (P_0y - P_Y[1])/2
  if pick == 2:
    P_0x = (P_0x - P_X[2])/2
    P_0y = (P_0y - P_Y[2])/2
#  print(P_0x,P_0y)
  string = '%i, %i, %f, %f\n' % (cnt,pick,P_0x,P_0y)
  f.write(string)
  cnt += 1
f.close()

cnt,pick,X,Y = np.genfromtxt('Serpinski_data.txt', delimiter = ',', comments = '#', unpack = True)
ax.plot(abs(X),abs(Y),'ro', ms=0.5)
```
